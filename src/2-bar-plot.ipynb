### Exercise 2: Bar Plot with Colors

```python
# TASK: Create a bar plot with the following data: categories = ['A', 'B', 'C', 'D'] and values = [5, 7, 3, 9].
# Use different colors for each bar and add a title to the plot.
import matplotlib.pyplot as plt
def plot_bar(categories, values):
    try:
        # Ensure inputs are lists
        if not isinstance(categories, list) or not isinstance(values, list):
            raise TypeError("Categories and values must be lists.")

        # Handle empty lists
        if not categories or not values:
            raise ValueError("Categories and values cannot be empty for a bar plot.")

        # Ensure equal length
        if len(categories) != len(values):
            raise ValueError("Categories and values must have the same length.")

        # Create bar plot
        plt.bar(categories, values)
        plt.title("Bar Plot")
        plt.xlabel("Categories")
        plt.ylabel("Values")
        plt.show()

    except Exception as e:
        print(f"Error: {e}")

# Example usage
categories = ['A', 'B', 'C', 'D']
values = [5, 3, 9, 6]
plot_bar(categories, values)
```
