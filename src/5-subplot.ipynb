### Exercise 5: Subplot Layout

```python
# TASK: Create a 2x2 subplot layout.
# Plot a line chart in the first subplot, a bar chart in the second, a scatter plot in the third, and a pie chart in the fourth.
import matplotlib.pyplot as plt
def create_subplots(x, y, categories, values, labels, sizes):
    try:
        # Ensure inputs are valid lists
        if not all(isinstance(lst, list) for lst in [x, y, categories, values, labels, sizes]):
            raise TypeError("All inputs must be lists.")

        # Check for empty lists
        if not x or not y:
            print("Warning: Empty data for line/scatter plot. Skipping these plots.")
        if not categories or not values:
            print("Warning: Empty data for bar plot. Skipping this plot.")
        if not labels or not sizes:
            print("Warning: Empty data for pie chart. Skipping this plot.")

        # Ensure equal lengths for necessary lists
        if len(x) != len(y):
            raise ValueError("x and y must be of equal length.")
        if len(categories) != len(values):
            raise ValueError("Categories and values must be of equal length.")
        if len(labels) != len(sizes):
            raise ValueError("Labels and sizes must be of equal length.")

        # Ensure non-negative pie chart sizes
        if any(size < 0 for size in sizes):
            raise ValueError("Pie chart sizes must be non-negative.")

        # Create subplots
        fig, axs = plt.subplots(2, 2, figsize=(10, 10))

        # First subplot: Line plot
        if x and y:
            axs[0, 0].plot(x, y, marker='o')
            axs[0, 0].set_title('Line Plot')

        # Second subplot: Bar plot
        if categories and values:
            axs[0, 1].bar(categories, values)
            axs[0, 1].set_title('Bar Plot')
        
        # Third subplot: Scatter plot
        if x and y:
            axs[1, 0].scatter(x, y)
            axs[1, 0].set_title('Scatter Plot')

        # Fourth subplot: Pie chart
        if labels and sizes:
            axs[1, 1].pie(sizes, labels=labels, autopct='%1.1f%%')
            axs[1, 1].set_title('Pie Chart')

        # Show plots
        plt.show()

    except Exception as e:
        print(f"Error: {e}")

# Example usage
x = [0, 1, 2, 3, 4]
y = [0, 1, 4, 9, 16]
categories = ['A', 'B', 'C', 'D']
values = [5, 3, 9, 6]
labels = ['Apple', 'Banana', 'Cherry', 'Date']
sizes = [10, 15, 7, 5]
create_subplots(x, y, categories, values, labels, sizes)
```
