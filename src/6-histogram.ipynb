### Exercise 6: Histogram Customization

```python
# TASK: Create a histogram for the following data: data = [1, 2, 2, 3, 3, 3, 4, 4, 4, 4, 5].
# Customize the histogram with a title, labels for the x-axis, and a specific color for the bars.
import matplotlib.pyplot as plt

# Function to create a histogram
def plot_histogram(data, bins=5):
    try:
        # Ensure input is a list
        if not isinstance(data, list):
            raise TypeError("Data must be a list.")

        # Handle empty list
        if not data:
            raise ValueError("Data cannot be empty for a histogram.")

        # Ensure bins is a positive integer
        if not isinstance(bins, int) or bins <= 0:
            raise ValueError("Bins must be a positive integer.")

        # Create histogram
        plt.hist(data, bins=bins, edgecolor='black')
        plt.title("Histogram")
        plt.xlabel("Values")
        plt.ylabel("Frequency")
        plt.show()

    except Exception as e:
        print(f"Error: {e}")

# Example usage
data = [1, 2, 2, 3, 3, 3, 4, 4, 4, 4, 5, 5, 6]
plot_histogram(data, bins=5)
```
