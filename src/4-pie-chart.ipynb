### Exercise 4: Pie Chart with Percentages

```python
# TASK: Create a pie chart with the following data: labels = ['Python', 'Java', 'C++', 'JavaScript'] and sizes = [40, 25, 20, 15].
# Display the percentages on the chart using autopct.
import matplotlib.pyplot as plt

def plot_pie(labels, sizes):
    try:
        # Ensure inputs are lists
        if not isinstance(labels, list) or not isinstance(sizes, list):
            raise TypeError("Labels and sizes must be lists.")

        # Handle empty lists
        if not labels or not sizes:
            raise ValueError("Labels and sizes cannot be empty for a pie chart.")

        # Ensure equal length
        if len(labels) != len(sizes):
            raise ValueError("Labels and sizes must have the same length.")

        # Ensure non-negative sizes
        if any(size < 0 for size in sizes):
            raise ValueError("Pie chart sizes must be non-negative.")

        # Create pie chart
        plt.pie(sizes, labels=labels, autopct='%1.1f%%')
        plt.title("Pie Chart")
        plt.show()

    except Exception as e:
        print(f"Error: {e}")

# Example usage
labels = ['Apple', 'Banana', 'Cherry', 'Date']
sizes = [10, 15, 7, 5]
plot_pie(labels, sizes)
```
