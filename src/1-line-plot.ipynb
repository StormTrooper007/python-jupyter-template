### Exercise 1: Line Plot Customization

```python
# TASK: Create a line plot with x values ranging from 0 to 10 and y values as the square of x.
# Customize the plot by adding a title, labels for both axes, and a grid.
import matplotlib.pyplot as plt

def plot_line(x, y):
    try:
        # Ensure inputs are lists
        if not isinstance(x, list) or not isinstance(y, list):
            raise TypeError("x and y must be lists.")

        # Handle empty lists
        if not x or not y:
            raise ValueError("x and y cannot be empty for a line plot.")

        # Ensure equal length
        if len(x) != len(y):
            raise ValueError("x and y must have the same length.")

        # Create line plot
        plt.plot(x, y, marker='o')
        plt.title("Line Plot")
        plt.xlabel("X-axis")
        plt.ylabel("Y-axis")
        plt.show()

    except Exception as e:
        print(f"Error: {e}")

# Example usage
x = [0, 1, 2, 3, 4]
y = [0, 1, 4, 9, 16]
plot_line(x, y)
```
